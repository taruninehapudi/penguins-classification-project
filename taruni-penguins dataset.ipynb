```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df['sex'].fillna('Male',inplace = True)
```

```python
df.info()
```

```python
df[['species','sex']].value_counts()
```

```python
mask = df['sex'].isna()
```

```python
df[mask]
```

```python
sns.countplot(x = 'species', data = df)
plt.title('Count of the total penguin species')
plt.show()
```

# sns.pairplot(df, hue = 'species')
plt.show()

```python
sns.boxplot(x='species', y='flipper_length_mm', data = df)
plt.title('Flipper length by species')
plt.show()
```

```python
sns.boxplot(x=df['bill_length_mm'])
plt.title('Boxplot of Bill Length (mm)')
plt.show()
```

```python
sns.boxplot(x=df['bill_depth_mm'])
plt.title('Boxplot of Bill Depth (mm)')
plt.show()
```

```python
sns.boxplot(x=df['flipper_length_mm'])
plt.title('Boxplot of Flipper Length (mm)')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y='flipper_length_mm', hue='species', data=df)
plt.title('Scatterplot between Bill length and Flipper Lengthb of species')
plt.show()
```

```python
sns.histplot(df['flipper_length_mm'], bins=20, kde=True)
plt.title('Histogram of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['flipper_length_mm'])  # Simple histogram
plt.title('Histogram of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
df.to_csv('df.csv', index=False)
```
